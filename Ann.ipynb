```
import io
import pandas as pd
bankdf = pd.read_csv(io.BytesIO(uploaded['bank.csv']))
bankdf.head()
```

```
## Feature Elimination

bankdf.drop(['RowNumber','CustomerId', 'Surname'], axis=1, inplace=True)
```

```
##Check for missing values in X

bankdf.isnull().sum()
```

```

X = bankdf.iloc[:,0:10].values
y = bankdf.iloc[:, 10].values

```

```
print(X[:10,:], '\n')
print(y[:10])
```

```

# convert Geography & Gender columnns into cateogrical data and also hot encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
```

```

#Gender
labelencoder1 = LabelEncoder()
X[:, 2] = labelencoder1.fit_transform(X[:, 2])


```

```

print(X[:10,:], '\n')
print(y[:10])
```

```
# Geography column. First create ColumnTrasfromer object
ct = ColumnTransformer([("encoder", OneHotEncoder(), [1])], remainder = 'passthrough')
#Once we have constructed this columnTransformer object, we have to fit and transform the dataset to label encode and one hot encode the column.
X = np.array(ct.fit_transform(X), dtype=np.str)
```

```
#print(X[:10,:], '\n')
#print(y[:10])
print(X[0])
```

```
from sklearn.model_selection import train_test_split
```

```
# Split Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=1)
```

```
# Normalization of data.
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
X_train = standardScaler.fit_transform(X_train)
X_test = standardScaler.transform(X_test)
```

```
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
# Create Model
import keras
import sys
from keras.models import Sequential #to initialize NN
from keras.layers import Dense #used to create layers in NN
```

```
classifier = Sequential()
#Input Layer
#Tip: units should be the average of nodes in the input layer (12 nodes) and the number of nodes in the output layer (1 node). For this case is 12+1/2 = 6#
classifier.add(Dense(units = 6, activation = 'relu', input_dim = 12 ))
# Hidden Layer
classifier.add(Dense(units = 6,  activation = 'relu'))
# output layer
classifier.add(Dense(units = 1,  activation = 'sigmoid'))
##sgd = optimizer.Adam(lr=0.001) 
classifier.compile(optimizer ='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100  )
```

```
# Prediction came to 86%
```

```
y_pred = classifier.predict(X_test)
```

```
#Threshold of 50%
y_pred = (y_pred > 0.5)
y_pred
```

```
# get Accuracy
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
