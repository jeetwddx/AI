### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women


1. There is no need to conduct statistical tests around this. Just compare the percentages and comment on the validity of the above statements.

2. you might want ot use the .sum(), .div() function here.
3. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

#### Conclusion:



```python
# get the distribution
user = pd.read_csv('user.csv')
ratings = pd.read_csv('Data.csv')
item = pd.read_csv('item.csv')
sex_dist = user.groupby(['gender']).count()
#sex_dist
# man are 3 times more than female( M: 670, F: 273) 2.45
# get Drama
item1 = item[['movie id','movie title','Drama', ]].copy()

onlydrama = item1['Drama'] > 0

drama = pd.merge(item1[onlydrama], ratings, on='movie id').merge(user, on='user id').copy()
drama1 = drama[['Drama','gender']]
drama1.groupby(['gender']).count()
#drama stats--->11008 Female vs 28887...now if we scale it by 2.45 i.e 26,969 Females watch Drama vs 28,887 men- True
#Now scifi
item2 = item[['movie id','movie title','Sci-Fi', ]].copy()
onlyscifi = item2['Sci-Fi'] > 0
scifi = pd.merge(item2[onlyscifi], ratings, on='movie id').merge(user, on='user id').copy()
scifi1 = scifi[['Sci-Fi','gender']]
scifi1.groupby(['gender']).count()
# Scifi stats 2626 females vs 10101 men...now we apply scaling of 2.45 6433 female vs 10101 men- False

#Now Romance
item3 = item[['movie id','movie title','Romance', ]].copy()
onlyromance = item3['Romance'] > 0
romance = pd.merge(item3[onlyromance], ratings, on='movie id').merge(user, on='user id').copy()
romance1 = romance[['Romance','gender']]
romance1.groupby(['gender']).count()
# Romance stats 5858 females vs 13603 men...now we apply scaling of 2.45 14352 female vs 13603 men- False
```
